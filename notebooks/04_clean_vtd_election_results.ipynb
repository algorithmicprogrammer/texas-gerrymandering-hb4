```python
from texas_gerrymandering_hb4.config import GEN_ELECTION_CSV
import pandas as pd

# Load General Election Results into pandas DataFrame
df = pd.read_csv(GEN_ELECTION_CSV, low_memory=False)

# Filter out U.S. Senate race
senate_df = df[df["Office"] == "U.S. Sen"].copy()

# Only columns of Interest will be kept
senate_df = senate_df[["County", "FIPS", "VTD", "cntyvtd", "vtdkeyvalue",
                       "Name", "Party", "Votes"]]

# Pivot so each candidate has their own column
senate_clean = senate_df.pivot_table(
    index=["County", "FIPS", "VTD", "cntyvtd", "vtdkeyvalue"],
    columns="Name",
    values="Votes",
    aggfunc="sum",
    fill_value=0
).reset_index()

# Renaming columns
senate_clean = senate_clean.rename_axis(None, axis=1)
senate_clean = senate_clean.rename(columns={
    "Cruz": "Cruz_R",
    "Allred": "Allred_D"
})

# Only show the votes of candidates Cruz and Allred
senate_two = senate_clean[["County", "FIPS", "VTD", "cntyvtd", "vtdkeyvalue",
                           "Cruz_R", "Allred_D"]]
```
