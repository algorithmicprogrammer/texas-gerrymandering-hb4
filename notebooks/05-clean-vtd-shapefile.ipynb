```python
from texas_gerrymandering_hb4.config import VTDS_SHP_FILE
import geopandas as gpd
from pathlib import Path

# Load shapefile
gdf = gpd.read_file(VTDS_SHP_FILE)

# Rename useful columns
gdf = gdf.rename(columns={
    "CNTY": "county_fips",
    "VTD": "vtd_code",
    "CNTYVTD": "precinct_id"
})

# Keep only useful fields
gdf = gdf[["county_fips", "vtd_code", "precinct_id", "geometry"]]

# Fix invalid polygons
gdf["geometry"] = gdf.buffer(0)

# Recalculate area/perimeter (in CRS units, metersÂ²)
gdf["area_m2"] = gdf.area
gdf["perimeter_m"] = gdf.length

# Save cleaned versions
gdf.to_file("data/interim/vtds_clean.gpkg", driver="GPKG", layer="vtds")
gdf.to_parquet("data/interim/vtds_clean.parquet", index=False)
```
